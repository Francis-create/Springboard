```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris

iris_extra = load_iris()

# determine which iris class it is.
# classification task
# based on the sepal/petal length/width

# Load iris.csv in DataFrame
iris = pd.read_csv('iris.csv')
# %% how many data points and features
print(iris.shape)

# %% What are the column names in our dataset
print(iris.columns)

# %% How many data points for each class are present
print(iris['species'].value_counts())


```

```python

# %% 2-D scatter plot
#iris.plot(kind='scatter', x = 'sepal_length', y = 'sepal_width')
plt.close('all')
plt.figure()
plt.scatter(iris['sepal_length'],iris['sepal_width'])
plt.grid(alpha = 0.5)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

```

```python
plt.figure()


# %% add colors
sns.set_style('whitegrid')
sns.FacetGrid(iris, hue='species', size = 4).map(plt.scatter,'sepal_length', 'sepal_width').add_legend()
```

```python
# %% 3D scatter plot

from mpl_toolkits.mplot3d import Axes3D
plt.figure()
ax = plt.subplot(111, projection='3d')

ax.scatter(iris['sepal_length'], iris['sepal_width'], iris['petal_length'], c = iris_extra['target'])

ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('Petal Length')
```
